Import Library

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

Import Dataset

```python
df =pd.read_csv("basket_analysis.csv")
df.head()
```

Data preprocessing

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.astype("int")
```

Training the datasets by Apriori model

```python
sup = apriori(df, min_support=0.1, use_colnames=True)
sup
```

```python
results = association_rules(sup, metric="lift", min_threshold=1)
results
```

Results sorted by descending confidence and lift

```python
results.sort_values(['confidence', "lift"], ascending = [False, False]).head(10)
```
