Import Pandas

```python
import pandas as pd
```

Import Dataset

```python
df = pd.read_csv("bread basket.csv")
```

```python
df
```

Data preprocessing

```python
grouped_data = df.groupby('Transaction')['Item'].apply(list).reset_index()
grouped_data
```

```python
#Create several columns
series = grouped_data['Item'].apply(pd.Series)
series
```

```python
#Stack columns
stack = series.stack()
stack
```

```python
#Encoding
encode = pd.get_dummies(stack)
encode
```

```python
df = encode.groupby(level=0).sum()
df
```

```python
df[df > 1] = 1
df
```

Training the datatsets by Apriori model

```python
from mlxtend.frequent_patterns import apriori, association_rules
support = apriori(df, min_support=0.02, use_colnames=True, max_len = 4)
support
```

```python
results = association_rules(support, metric="lift", min_threshold=1)
results
```

Results sorted by descending lift

```python
results.sort_values(by='lift', ascending=False).head(10)
```
