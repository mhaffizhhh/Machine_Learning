Import Pandas

```python
import pandas as pd
```

Import Dataset

```python
df = pd.read_excel("Assignment-1_Data.xlsx")
df
```

Data preprocessing

```python
df1 = pd.pivot_table(index="BillNo", columns="Itemname", data=df, values = "Price")
```

```python
df1.fillna(0, inplace=True)
```

```python
df1[df1 != 0] = 1
df1
```

Training the datasets by Apriori model

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
sup = apriori(df1, min_support=0.03, use_colnames=True, max_len=2)
sup
```

```python
results = association_rules(sup, metric="lift", min_threshold=3)
results
```

Results sorted by descending lift

```python
results.sort_values("lift", ascending=False).head(10)
```
