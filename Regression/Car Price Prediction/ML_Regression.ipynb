Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import Dataset

```python
df = pd.read_csv("car_price.csv")
df
```

Data Preprocessing

```python
def convert_price(price):
    price = price.replace(',', '').strip()  # Menghilangkan koma dan spasi
    if 'Lakh' in price:
        return float(price.replace(' Lakh', '')) * 100000
    elif 'Crore' in price:
        return float(price.replace(' Crore', '')) * 10000000
    else:
        return float(price)
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].apply(convert_price)
```

```python
df["kms_driven"] = df["kms_driven"].str.split(" ").str.get(0).str.replace(",","").astype(int)
```

```python
df["engine"] = df["engine"].str.split(" ").str.get(0).astype(int)
```

```python
df["Seats"] = df["Seats"].str.split(" ").str.get(0).astype(int)
```

SVR

```python
y = np.array(y).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
from sklearn.svm import SVR
regressor = SVR(kernel = "rbf")
regressor.fit(X_train, y_train)
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
