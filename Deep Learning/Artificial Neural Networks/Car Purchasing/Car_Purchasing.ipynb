Import Library

```python
import tensorflow as tf
import pandas as pd
```

Import Dataset

```python
df = pd.read_csv("car_purchasing.csv", encoding='latin-1')
df
```

```python
X = df.drop(columns=['customer name', 'customer e-mail', 'country', 'car purchase amount'])
y = df['car purchase amount']
```

Split Dataset Into Training and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.values.reshape(-1, 1))
y_test = sc_y.transform(y_test.values.reshape(-1, 1))
```

Create ANN

```python
ann = tf.keras.models.Sequential()
```

Add Input Layer and First Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

Add Second Hidden Layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

Add Output Layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

Compile ANN

```python
ann.compile(optimizer = "adam", loss="mean_squared_error")
```

Train The ANN

```python
ann.fit(X_train, y_train, batch_size = 25, epochs = 100)
```

Predict Test Set

```python
y_pred = ann.predict(X_test, batch_size = 25)
```

Evaluate ANN

```python
from sklearn.metrics import mean_squared_error, r2_score
print("R2 =",r2_score(y_test, y_pred))
print("MSE =",mean_squared_error(y_test, y_pred))
```

```python
import numpy as np
y_pred = sc_y.inverse_transform(y_pred)
y_test = sc_y.inverse_transform(y_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
